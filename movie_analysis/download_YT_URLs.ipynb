```python
MovieTitle = "Beautiful Boy" # test
#Takes in movie title, then appends 'trailer' and searches on YT. Returns URL
#code to use if using header: , headers = header, timeout=4
def take_first_trailer(MovieTitle):
    TextToSearch = MovieTitle + ' trailer' #get weird first result if there are spaces
    TextToSearch.replace(" ", "")
    TextToSearch.strip()
    query = TextToSearch
    url = "https://www.youtube.com/results?search_query=" + query
    #try:
    response = requests.get(url)
    html = response.text
    soup = bs(html, 'html.parser')
    ListOfVideos = []
    for vid in soup.findAll(attrs={'class':'yt-uix-tile-link'}):
        link = 'https://www.youtube.com' + vid['href']
        ListOfVideos.append(link)
        
        #print(ListOfVideos[1]) #first element is garbage so print second one
    sleep(randint(1,10))
    return ListOfVideos[1]
    #except:
        #print("Google plz let me finish my homework")
    
#take_first_trailer(MovieTitle) #test 
```
